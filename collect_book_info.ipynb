```python
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from utils import *
```

```python
books_read_in_2023 = pd.read_csv('data/goodreads_read_parsed.csv')
```

```python
books_read_in_2023.head()
```

```python
# Subset the DataFrame to include only the desired columns
subset_df = books_read_in_2023[['Book Id', 'Title', 'Author', 'My Rating', 'Average Rating', 'Number of Pages','Year Published','Date Added', 'Date Read','My Review']]
```

```python
# Convert the subset DataFrame to a dictionary
# 'records' mode will make each row a dictionary in a list
books_dict = subset_df.to_dict(orient='records')
```

```python
# Convert the dictionary to a JSON string
# Write personal reading info to JSON
books_json = json.dumps(books_dict, indent=4)
with open('data/book_reading_info.json', 'w') as file:
    file.write(books_json)
```

```python
# Scrape each book and write to JSON
with open('data/book_scrape_info.json', 'w') as file:
    for i in range(len(books_read_in_2023)):
        book = books_read_in_2023['Book Id'].iloc[i]
        book_dict = scrape_book(str(book))
        books_json = json.dumps(book_dict, indent=4)
        file.write(books_json)
```
