```python
import pandas as pd
import numpy as np
import os
from utils import *
```

```python
# Import library export data
data = pd.read_csv('data/goodreads_library_export.csv')
```

```python
# Convert 'Date Read' column to datetime
data['Date Read'] = pd.to_datetime(data['Date Read'], format='%Y/%m/%d')

# Filter books read in 2023
books_read_in_2023 = data[data['Date Read'].dt.year == 2023]
# Sort by date read 
books_read_in_2023.sort_values(by='Date Read',ascending=False, inplace=True)
```

```python
user_id = os.environ.get('USERID')

# Scrape 'Date Started' info since it's not included in the export
dates1 = get_dates(user_id,'1')
dates2 = get_dates(user_id,'2')
dates2=dates2[0:17] 
dates = dates1 + dates2
parsed_dates=pd.to_datetime(dates)
formatted_dates = parsed_dates.strftime('%Y-%m-%d')
```

```python
# Add as new column
books_read_in_2023['Date Added'] = formatted_dates
```
